This notebook describes how to implement ResNet50 model from paper Deep Residual Learning for Image Recognition arXiv:1512.03385 [cs.CV] in Tensorflow Keras.

## The problem of very deep neural networks

The main benefit of a very deep network is that it can represent very complex functions. It can also learn features at many different levels of abstraction, from edges (at the lower layers) to very complex features (at the deeper layers). However, using a deeper network doesn't always help. A huge barrier to training them is vanishing gradients: very deep networks often have a gradient signal that goes to zero quickly, thus making gradient descent unbearably slow. More specifically, during gradient descent, as you backprop from the final layer back to the first layer, you are multiplying by the weight matrix on each step, and thus the gradient can decrease exponentially quickly to zero (or, in rare cases, grow exponentially quickly and "explode" to take very large values). 

During training, you might therefore see the magnitude (or norm) of the gradient for the earlier layers descrease to zero very rapidly as training proceeds: 

<center><img src="images/vanishing_grad.png" style="width:450px;height:220px;"/></center>
<center><b>Figure 1</b> : <b>Vanishing gradient</b> <br> The speed of learning decreases very rapidly for the early layers as the network trains</center>

In Resnet, skip connection or shortcut allows gradient to be directly propagated to earlier layers. The resnet blocks can learn identity function so we can add more layers to form a deep network without harming the network performance.

## ResNet blocks
There are 2 types of ResNet blocks dependending main on whether input/output dimensions are same or different.

### 1. Identity block
<center><img src="images/id_block.png" style="width:650px;height:150px;"></center>
<center><b>Figure 2</b> : <b>Identity block</b> <br> Skip connection skips over 3 layes</center>

### 2. Convolutional block
<center><img src="images/conv_block.png" style="width:650px;height:200px;"></center>
<center><b>Figure 2</b> : <b>Convolutional block</b> <br> Conv2D layers on shortcut path to downsample input size</center>

```python
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, add, Input, MaxPool2D, ZeroPadding2D, GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model
```

```python
def conv_block(X, F, stage, block, f=3, s=2):
    #filters for 3 blocks
    F1, F2, F3 = F

    X_shortcut = X
    
    base_name = 'conv_stage'+str(stage)+'_block'+str(block)

    #block1
    X = Conv2D(filters=F1, kernel_size=(1, 1), strides=(s, s), padding='valid', name=base_name+'_'+'conv1_main')(X)
    X = BatchNormalization(axis=3, name=base_name+'_'+'bn1_main')(X)
    X = Activation('relu', name=base_name+'_'+'relu1_main')(X)
    #block2
    X = Conv2D(filters=F2, kernel_size=(f, f), strides=(1, 1), padding='same', name=base_name+'_'+'conv2_main')(X)
    X = BatchNormalization(axis=3, name=base_name+'_'+'bn2_main')(X)
    X = Activation('relu', name=base_name+'_'+'relu2_main')(X)
    #block3
    X = Conv2D(filters=F3, kernel_size=(1, 1), strides=(1, 1), padding='valid', name=base_name+'_'+'conv3_main')(X)
    X = BatchNormalization(axis=3, name=base_name+'_'+'bn3_main')(X)

    #add conv layer to shortcut
    X_shortcut = Conv2D(filters=F3, kernel_size=(1, 1), strides=(s, s), padding='valid', name=base_name+'_'+'conv0_shortcut')(X_shortcut)
    X_shortcut = BatchNormalization(axis=3, name=base_name+'_'+'bn0_shortcut')(X_shortcut)

    #add shortcut to output of main path
    X = add([X_shortcut, X])

    #Add activation
    X = Activation('relu', name=base_name+'_relu_output')(X)

    return X

def identity_block(X, F, stage, block, f=3):
    #filters for 3 blocks
    F1, F2, F3 = F
    
    #save the input for later use
    X_shortcut = X
    
    #basename for layers
    base_name = 'identity_stage'+str(stage)+'_block'+str(block)

    #block1
    X = Conv2D(filters=F1, kernel_size=(1, 1), padding='valid', name=base_name+'_'+'conv1_main')(X)
    X = BatchNormalization(axis=3, name=base_name+'_'+'bn1_main')(X)
    X = Activation('relu', name=base_name+'_'+'relu1_main')(X)
    #block2
    X = Conv2D(filters=F2, kernel_size=(f, f), padding='same', name=base_name+'_'+'conv2_main')(X)
    X = BatchNormalization(axis=3, name=base_name+'_'+'bn2_main')(X)
    X = Activation('relu', name=base_name+'_'+'relu2_main')(X)
    #block3
    X = Conv2D(filters=F3, kernel_size=(1, 1), padding='valid', name=base_name+'_'+'conv3_main')(X)
    X = BatchNormalization(axis=3, name=base_name+'_'+'bn3_main')(X)

    #add shortcut to output of main path
    X = add([X, X_shortcut])

    #Add activation
    X = Activation('relu')(X)

    return X

def ResNet50(input_shape=(224, 224, 3), num_classes=1000, weights_path=None):
    #Create input tensor
    input_img = Input(shape=input_shape, name='input')

    #Stage1
    #Zeropadding + Conv + BN + Relu
    X = ZeroPadding2D(padding=(3, 3))(input_img)
    X = Conv2D(filters=64, kernel_size=(7, 7), strides=(2, 2), padding='valid', name='conv1')(X)
    X = BatchNormalization(axis=3, name='bn1')(X)
    X = Activation('relu')(X)

    #Pad zeros to adjust the input image size
    X = ZeroPadding2D(padding=(1, 1))(X)
    
    #Maxpooling
    X = MaxPool2D(pool_size=(3, 3), strides=(2, 2), name='pool1')(X)

    #Stage2
    X = conv_block(X, F=(64, 64, 256), stage=2, block=1, f=3, s=1)
    X = identity_block(X, F=(64, 64, 256), stage=2, block=2, f=3)
    X = identity_block(X, F=(64, 64, 256), stage=2, block=3, f=3)

    #Stage3
    X = conv_block(X, F=[128, 128, 512], stage=3, block=1, f=3, s=2)
    X = identity_block(X, F=(128, 128, 512), stage=3, block=2, f=3)
    X = identity_block(X, F=(128, 128, 512), stage=3, block=3, f=3)
    X = identity_block(X, F=(128, 128, 512), stage=3, block=4, f=3)

    #Stage4
    X = conv_block(X, F=(256, 256, 1024), stage=4, block=1, f=3, s=2)
    X = identity_block(X, F=(256, 256, 1024), stage=4, block=2, f=3)
    X = identity_block(X, F=(256, 256, 1024), stage=4, block=3, f=3)
    X = identity_block(X, F=(256, 256, 1024), stage=4, block=4, f=3)
    X = identity_block(X, F=(256, 256, 1024), stage=4, block=5, f=3)
    X = identity_block(X, F=(256, 256, 1024), stage=4, block=6, f=3)

    #Stage5
    X = conv_block(X, F=(512, 512, 2048), stage=5, block=1, f=3, s=2)
    X = identity_block(X, F=(512, 512, 2048), stage=5, block=2, f=3)
    X = identity_block(X, F=(512, 512, 2048), stage=5, block=3, f=3)

    #Average pooling
    X = GlobalAveragePooling2D(name='avg_pool')(X)

    #FC layer
    X = Dense(num_classes, activation='softmax', name='fc1000')(X)

    model = Model(inputs=input_img, outputs=X)

    #load weights if weights file present
    if weights_path:
        model.load_weights(weights_path)

    return model    
```

```python
resnet_model = ResNet50()
resnet_model.summary()
```

```python
import os
import requests

weights_filename = "resnet50_weights_tf_dim_ordering_tf_kernels.h5"

weights_dir = os.path.join(os.path.expanduser("~"), ".keras", "models")

weigths_path = os.path.join(weights_dir, weights_filename)

#download the imagenet weights if not already in downloaded
if not os.path.exists(weigths_path):
    url = "https://github.com/keras-team/keras-applications/releases/download/resnet/"
    r = requests.get(url+weights_filename, allow_redirects=True)
    with open(weigths_path, 'wb') as file:
        file.write(r.content)  
```

```python
#load the pretrained model weights
resnet_model.load_weights(weigths_path)
```

```python
import numpy as np
import requests
import cv2

def url_to_image(url):
	# download the image, convert it to a numpy array, and then read
	# it into OpenCV format
    r = requests.get(url, allow_redirects=True)
    image = np.asarray(bytearray(r.content), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
	# return the image
    return image
```

```python
import matplotlib.pyplot as plt

def predict_from_url(model, url):
    img = url_to_image(url)
    img = img[:, :, ::-1]
    plt.imshow(img)
    plt.show()

    x = preprocess_input(img)
    predictions = model.predict(x)[0]
    print("Top 3 predictions")
    print(get_topk_predictions(predictions))
```
